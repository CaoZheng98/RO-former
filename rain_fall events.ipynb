```python
#----------------------------------------------筛选条件------------------------------------------------------------
RI = 10
OI = 50
rain_events = [event for event in rain_events if event['Rainfall intensity'] > RI]
filtered_events = [event for event in rain_events if event['outfall intensity'] > OI ]
print(len(filtered_events))
```

```python
[rainfall_event['Rainfall intensity'] for rainfall_event  in filtered_events]
```

```python
import matplotlib.pyplot as plt

# 假设 idx 已经定义，并且 filtered_events[idx]['data_frame'] 包含了所需的数据
idx = 8
src = filtered_events[idx]['data_frame']['实时降雨量(mm/5min)'].tolist()
tgt = filtered_events[idx]['data_frame']['实时流量(m³/5min)'].tolist()

# 创建时间序列（假设数据是按时间顺序排列的）
time = list(range(len(src)))

# 创建图形和主坐标轴
fig, ax1 = plt.subplots(figsize=(10, 6))

# 在主坐标轴上绘制实时降雨量
ax1.set_xlabel('时间序列')
ax1.set_ylabel('实时降雨量(mm/5min)', color='blue')
ax1.plot(time, src, label='实时降雨量(mm/5min)', color='blue', marker='o')
ax1.tick_params(axis='y', labelcolor='blue')

# 创建第二个坐标轴
ax2 = ax1.twinx()  # 创建共享x轴的第二个y轴
ax2.set_ylabel('实时流量(m³/5min)', color='red')
ax2.plot(time, tgt, label='实时流量(m³/5min)', color='red', marker='x')
ax2.tick_params(axis='y', labelcolor='red')

# 添加标题和图例
# plt.title('实时降雨量与实时流量的比较')
ax1.legend(loc='upper left')  # 为第一个坐标轴添加图例
ax2.legend(loc='upper right')  # 为第二个坐标轴添加图例

# 显示网格
ax1.grid(True)

# 显示图表
plt.show()
```
