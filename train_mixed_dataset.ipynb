```python
from models.transformer.transformer import Transformer
import torch
max_len = 2000
pad_idx = 1
d_model = 64
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = Transformer(src_len_max = max_len,trg_len_max=max_len, src_pad_idx = pad_idx, trg_pad_idx=pad_idx, d_src=1, d_trg=1, d_model=d_model, trg_size=1).to(device)
```

```python
#DATASET INCLUDE ALL DATA
from data.dataset import DT_Dataset_all
from torch.utils.data import DataLoader
from utils.collate_fn import collate_fn
# _________dataset________
out_fall_name = 'JHHHHC'
max_patience = 12
batch_size = 100
ds_DT = DT_Dataset_all(outfall_name = out_fall_name,max_patience=max_patience)
ds_val = DT_Dataset_all(outfall_name = out_fall_name,max_patience=max_patience,type='val')
val_loader = DataLoader(dataset=ds_val, batch_size=batch_size, shuffle=False,collate_fn=collate_fn)
```

```python
#DATASET INCLUDE PRETRAIN DATA
from data.dataset import pre_train_Dataset
out_fall_name = 'JHHN2'
ds_pretrain = pre_train_Dataset(outfall_name=out_fall_name)
```

```python
from data.dataset import Dataset_MIXED
# 创建混合数据集
mixed_dataset = Dataset_MIXED(ds_DT, ds_pretrain)
train_loader = DataLoader(dataset=mixed_dataset, batch_size=batch_size, shuffle=True,collate_fn=collate_fn)
```

```python
from utils.train_full import train_full
from pathlib import Path
from torch import nn
from utils.metrics import loss_func_nse,loss_func_mse,loss_func_NZ,loss_func_NSE_enhanced,loss_func_advanced
from utils.lr_strategies import SchedulerFactory
bos_len = 3
learning_rate = 1e-2
n_epochs = 500
loss_f = loss_func_NZ
Rainfall_intensity = 'NONE'
saving_root = Path("./runs") / f'TD_ml_{max_len}_padidx_{pad_idx}_bos_len_{bos_len}_lr_{learning_rate}_epochs_{n_epochs}_d_model_{d_model}_of_{out_fall_name}_RI_{Rainfall_intensity}_LF_{loss_f.__name__}_DATASET_f'
# 如果有多个 GPU，使用 DataParallel 包装模型
if torch.cuda.device_count() > 1:
    print(f"Using {torch.cuda.device_count()} GPUs")
    model = nn.DataParallel(model)
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
scheduler_paras = {"scheduler_type": "warm_up", "warm_up_epochs": n_epochs * 0.25, "decay_rate":0.99}
scheduler = SchedulerFactory.get_scheduler(optimizer, **scheduler_paras)
train_full(model, train_loader, val_loader, optimizer,bos_len,scheduler, loss_f, n_epochs, device, saving_root)
```
